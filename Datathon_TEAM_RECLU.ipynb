```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# --- 1) Carga tu DataFrame `df2` ---
# df2 = pd.read_csv("cobranza_history.csv")

# Convertir fechas a tipo datetime
df2['fechaCobroBanco'] = pd.to_datetime(df2['fechaCobroBanco'], errors='coerce')
df2['fechaEnvioCobro'] = pd.to_datetime(df2['fechaEnvioCobro'], errors='coerce')

# 2) Ingeniería de variables mínima

# 2.1) Indicador de pago en el intento
df2['paid_any'] = (df2['montoCobrado'] > 0).astype(int)

# 2.2) Monto extraído considerando domiciliación parcial BBVA
def calc_amount_extracted(row):
    if row['paid_any'] == 1:
        if row['idBanco'] == 'BBVA':
            return min(row['montoCobrar'], row['montoCobrado'])
        else:
            return row['montoCobrado']
    else:
        return 0

df2['amount_extracted'] = df2.apply(calc_amount_extracted, axis=1)

# 2.3) Tiempo de respuesta banco (horas)
df2['hours_to_response'] = (df2['fechaCobroBanco'] - df2['fechaEnvioCobro']).dt.total_seconds() / 3600

# 3) Definición de features y preprocesador
num_feats = ['consecutivoCobro', 'hours_to_response']
cat_feats = ['idBanco', 'TipoEnvio']

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_feats),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_feats),
])

# 4) Pipeline de Clasificación: probabilidad de pago
clf_pipe = Pipeline([
    ('pre', preprocessor),
    ('clf', RandomForestClassifier(random_state=42))
])

clf_param_grid = {
    'clf__n_estimators': [50, 100],
    'clf__max_depth': [5, 10, None],
}

clf_grid = GridSearchCV(
    clf_pipe,
    clf_param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)

X_clf = df2[num_feats + cat_feats]
y_clf = df2['paid_any']
clf_grid.fit(X_clf, y_clf)

# 5) Pipeline de Regresión: monto extraído
df2_paid = df2[df2['paid_any'] == 1].copy()

reg_pipe = Pipeline([
    ('pre', preprocessor),
    ('reg', RandomForestRegressor(random_state=42))
])

reg_param_grid = {
    'reg__n_estimators': [50, 100],
    'reg__max_depth': [5, 10, None],
}

reg_grid = GridSearchCV(
    reg_pipe,
    reg_param_grid,
    cv=5,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=1
)

X_reg = df2_paid[num_feats + cat_feats]
y_reg = df2_paid['amount_extracted']
reg_grid.fit(X_reg, y_reg)

# 6) Cálculo de EB por intento
cost_map = {
    ('BBVA', 'Tradicional'): 8.00, ('BBVA', 'Interbancaria'): 6.00,
    ('BBVA', 'Parcial'): 1.60, ('BBVA', 'Matutino'): 8.00,
    ('BBVA', 'En Línea'): 4.00,
    ('Banamex', 'Tradicional'): 1.75, ('Banamex', 'Interbancaria'): 1.75,
    ('Banorte', 'Tradicional'): 2.50, ('Banorte', 'Interbancaria'): 4.50,
    ('Santander', 'Tradicional'): 2.82, ('Santander', 'Interbancaria'): 3.18
}

df2['p_pred'] = clf_grid.predict_proba(X_clf)[:, 1]
df2['amt_pred'] = reg_grid.predict(preprocessor.transform(X_clf))
df2['cost'] = df2.apply(lambda r: cost_map.get((r['idBanco'], r['TipoEnvio']), 0.0), axis=1)
df2['EB'] = df2['p_pred'] * df2['amt_pred'] - df2['cost']

# 7) Agregar EB por crédito o lista de cobro
eb_by_credit = df2.groupby('idCredito')['EB'].sum().reset_index().rename(columns={'EB': 'EB_total'})

import ace_tools as tools
tools.display_dataframe_to_user("EB por crédito", eb_by_credit)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from scipy.optimize import minimize

# --- 1. Cargar dataset limpio ---
df = pd.read_csv("/content/df_final_limpio1206.csv", low_memory=False)

# --- 2. Parsear fechas y horas necesarias ---
df['fechaCreacionLista'] = pd.to_datetime(df['fechaCreacionLista'], errors='coerce', dayfirst=True)
df['horaCreacionLista'] = pd.to_datetime(df['horaCreacionLista'], format='%H:%M:%S', errors='coerce')
df['fechaEnvioCobro'] = pd.to_datetime(df['fechaEnvioCobro'], errors='coerce', dayfirst=True)
df['fechaCobroBanco'] = pd.to_datetime(df['fechaCobroBanco'], errors='coerce', dayfirst=True)

# --- 3. Calcular delay en días ---
df['fechaHoraCreacion'] = df['fechaCreacionLista'].dt.normalize() + pd.to_timedelta(
    df['horaCreacionLista'].dt.hour.fillna(0).astype(int), unit='h'
) + pd.to_timedelta(
    df['horaCreacionLista'].dt.minute.fillna(0).astype(int), unit='m'
) + pd.to_timedelta(
    df['horaCreacionLista'].dt.second.fillna(0).astype(int), unit='s'
)

df['delay_days'] = (df['fechaEnvioCobro'] - df['fechaHoraCreacion']).dt.total_seconds() / 86400

# --- 4. Probabilidad de cobro (simulada o del modelo previo)
df['p_pred'] = (df['montoCobrado'] > 0).astype(int)  # puedes reemplazar esto por pipeline.predict_proba si ya lo tienes

# --- 5. Costo: BBVA solo cobra si es exitoso
bbva_ids = [2]  # idBanco para BBVA
df['es_exitoso'] = df['montoCobrado'] > 0

cost_map = {
    (2, 'TRADICIONAL'): 8.00, (2, 'INTERBANCARIO'): 6.00, (2, 'PARCIAL'): 1.60,
    (2, 'MATUTINO'): 8.00, (2, 'EN LINEA'): 4.00,
    (12, 'TRADICIONAL'): 1.75, (12, 'INTERBANCARIO'): 1.75,
    (14, 'TRADICIONAL'): 2.50, (14, 'INTERBANCARIO'): 4.50,
    (72, 'TRADICIONAL'): 2.82, (72, 'INTERBANCARIO'): 3.18
}

df['costo'] = df.apply(
    lambda r: cost_map.get((r['idBanco'], r['TipoEnvio']), 0.0) if r['idBanco'] not in bbva_ids
    else (cost_map.get((r['idBanco'], r['TipoEnvio']), 0.0) if r['es_exitoso'] else 0.0),
    axis=1
)

# --- 6. Filtrar primeros 5 intentos ---
df = df[df['consecutivoCobro'] <= 5].copy()

# --- 7. Agrupar por combinación de estrategia (emisora + canal) ---
estrategias_data = df.groupby(['idEmisora', 'TipoEnvio']).agg(
    costo_promedio=('costo', 'mean'),
    conversion_promedio=('p_pred', 'mean'),
    tiempo_promedio=('delay_days', 'mean'),
    intentos=('p_pred', 'count')
).reset_index()

# --- 8. Normalizar para combinación multiobjetivo ---
scaler = MinMaxScaler()
scaled = scaler.fit_transform(estrategias_data[['costo_promedio', 'tiempo_promedio', 'conversion_promedio']])
estrategias_data[['costo_norm', 'tiempo_norm', 'conv_norm']] = scaled

# --- 9. Definir función objetivo (min costo + tiempo, max conversion) ---
def objective_mix(weights):
    costo = np.dot(weights, estrategias_data['costo_norm'])
    tiempo = np.dot(weights, estrategias_data['tiempo_norm'])
    conversion = np.dot(weights, estrategias_data['conv_norm'])
    return costo + tiempo - conversion

# --- 10. Restricciones de mezcla ---
n = estrategias_data.shape[0]
constraints = [{'type': 'eq', 'fun': lambda w: np.sum(w) - 1}]
bounds = [(0, 1)] * n
initial = np.ones(n) / n

result = minimize(objective_mix, initial, bounds=bounds, constraints=constraints)

# --- 11. Mezcla óptima
estrategias_data['peso_optimo'] = result.x
estrategias_data = estrategias_data.sort_values('peso_optimo', ascending=False)

# --- 12. Mostrar top estrategias
print("\n📊 Mix óptimo multiobjetivo (solo primeros 5 intentos por crédito):")
print(estrategias_data[['idEmisora', 'TipoEnvio', 'peso_optimo', 'costo_promedio', 'tiempo_promedio', 'conversion_promedio']].head(10))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import classification_report

# --- 1. Cargar datos
df = pd.read_csv("/content/df_final_limpio1206.csv", low_memory=False)

# --- 2. Asegurar fechas y horas necesarias
df['fechaCreacionLista'] = pd.to_datetime(df['fechaCreacionLista'], errors='coerce', dayfirst=True)
df['horaCreacionLista'] = pd.to_datetime(df['horaCreacionLista'], format='%H:%M:%S', errors='coerce')
df['fechaEnvioCobro'] = pd.to_datetime(df['fechaEnvioCobro'], errors='coerce', dayfirst=True)

df['fechaHoraCreacion'] = df['fechaCreacionLista'].dt.normalize() + pd.to_timedelta(
    df['horaCreacionLista'].dt.hour.fillna(0).astype(int), unit='h'
) + pd.to_timedelta(
    df['horaCreacionLista'].dt.minute.fillna(0).astype(int), unit='m'
) + pd.to_timedelta(
    df['horaCreacionLista'].dt.second.fillna(0).astype(int), unit='s'
)

df['hour'] = df['fechaHoraCreacion'].dt.hour
df['dow'] = df['fechaHoraCreacion'].dt.weekday
df['delay_days'] = (df['fechaEnvioCobro'] - df['fechaHoraCreacion']).dt.total_seconds() / (3600 * 24)

# --- 3. Crear columna bajo_riesgo
# Agrupamos para saber si pagó antes del intento 4
df['es_exitoso'] = df['montoCobrado'] > 0

primeros_intentos = df[df['consecutivoCobro'] <= 3]
primer_pago = primeros_intentos[primeros_intentos['es_exitoso']].groupby('idCredito')['consecutivoCobro'].min()

df['bajo_riesgo'] = df['idCredito'].isin(primer_pago.index)

# --- 4. Features
features = ['idBanco', 'TipoEnvio', 'idEmisora', 'consecutivoCobro', 'montoCobrar', 'hour', 'dow', 'delay_days']
target = 'bajo_riesgo'
num_features = ['consecutivoCobro', 'montoCobrar', 'hour', 'dow', 'delay_days']
cat_features = ['idBanco', 'TipoEnvio', 'idEmisora']

# --- 5. Preprocesamiento y entrenamiento
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)
])

pipeline = Pipeline([
    ('pre', preprocessor),
    ('clf', LogisticRegression(max_iter=1000))
])

df_model = df[features + [target]].dropna()
X = df_model[features]
y = df_model[target].astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
pipeline.fit(X_train, y_train)

# --- 6. Evaluación
y_pred = pipeline.predict(X_test)
print(classification_report(y_test, y_pred, target_names=['alto_riesgo', 'bajo_riesgo']))
```
