```python
import pandas as pd

archivos = [
    'ListaCobroDetalle2022.csv',
    'ListaCobroDetalle2023.csv',
    'ListaCobroDetalle2024.csv',
    'ListaCobroDetalle2025.csv',
]

df_total = pd.concat([pd.read_csv(archivo) for archivo in archivos], ignore_index=True)
```

```python
df_total
```

```python
df2 = pd.read_csv('ListaCobro.csv', dtype=str)
```

```python
# Asegurar que la clave esté como string en ambos DataFrames
df_total['idListaCobro'] = df_total['idListaCobro'].astype(str)
df2['idListaCobro'] = df2['idListaCobro'].astype(str)

# Realizar el merge
merged_df = df_total.merge(df2, on='idListaCobro', how='left')

# Verificar las primeras filas
merged_df.head()
```

```python
merged_df
```

```python
merged_df.drop('idBanco_y', axis=1, inplace=True)
merged_df.rename(columns={'idBanco_x': 'idBanco'}, inplace=True)
```

```python
merged_df
```

```python
# Asegúrate de convertir la columna a tipo datetime
merged_df['fechaCreacionLista'] = pd.to_datetime(merged_df['fechaCreacionLista'], dayfirst=True, errors='coerce')

# Crear nueva columna con solo la hora
merged_df['horaCreacionLista'] = merged_df['fechaCreacionLista'].dt.time
```

```python
# 1. Crear nueva columna con la hora (ya lo habías hecho antes, por si acaso)
merged_df['horaCreacionLista'] = merged_df['fechaCreacionLista'].dt.time

# 2. Reemplazar la columna original por solo la fecha
merged_df['fechaCreacionLista'] = merged_df['fechaCreacionLista'].dt.date

# 3. Reordenar columnas para que 'fechaCreacionLista' y 'horaCreacionLista' estén juntas
# Obtener todas las columnas
cols = merged_df.columns.tolist()

# Mover 'horaCreacionLista' justo después de 'fechaCreacionLista'
cols.remove('horaCreacionLista')
fecha_idx = cols.index('fechaCreacionLista')
cols.insert(fecha_idx + 1, 'horaCreacionLista')

# Reordenar el DataFrame
merged_df = merged_df[cols]
```

```python
merged_df
```

```python
# Crear un nuevo DataFrame solo con los 4 bancos más frecuentes
primeros_bancos = [12, 2, 14, 72]

df_bancos = merged_df[merged_df['idBanco'].astype(int).isin(primeros_bancos)]

# Verificamos los resultados
df_bancos['idBanco'].value_counts()
```
